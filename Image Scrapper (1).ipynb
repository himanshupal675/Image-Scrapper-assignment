## Q1. Write a python program to extract the video URL of the first five videos.

````python
import requests
from bs4 import BeautifulSoup

def extract_video_urls(query, num_videos=5):
    # Example YouTube search URL
    search_url = f'https://www.youtube.com/results?search_query={query}'

    # Send a GET request to the search URL
    response = requests.get(search_url)

    if response.status_code == 200:
        # Parse HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the elements containing video URLs
        video_elements = soup.select('.yt-simple-endpoint.style-scope.ytd-video-renderer')

        # Extract video URLs from the elements
        video_urls = [element['href'] for element in video_elements[:num_videos]]

        return video_urls

    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

# Example usage: Extract video URLs for the query 'python tutorial'
video_urls = extract_video_urls('python tutorial', num_videos=5)

if video_urls:
    print("First five video URLs:")
    for idx, url in enumerate(video_urls, start=1):
        print(f"{idx}. https://www.youtube.com{url}")
````

## Q2. Write a python program to extract the URL of the video thumbnails of the first five videos.

## Q3. Write a python program to extract the title of the first five videos.

````python
import requests
from bs4 import BeautifulSoup

def extract_video_titles(query, num_titles=5):
    # Example YouTube search URL
    search_url = f'https://www.youtube.com/results?search_query={query}'

    # Send a GET request to the search URL
    response = requests.get(search_url)

    if response.status_code == 200:
        # Parse HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the elements containing video titles
        title_elements = soup.select('.yt-simple-endpoint.style-scope.ytd-video-renderer h3')

        # Extract video titles from the elements
        video_titles = [element.get_text() for element in title_elements[:num_titles]]

        return video_titles

    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

# Example usage: Extract video titles for the query 'python tutorial'
video_titles = extract_video_titles('python tutorial', num_titles=5)

if video_titles:
    print("First five video titles:")
    for idx, title in enumerate(video_titles, start=1):
        print(f"{idx}. {title}")
````

## Q4. Write a python program to extract the number of views of the first five videos.

````python
import requests
from bs4 import BeautifulSoup

def extract_video_views(query, num_views=5):
    # Example YouTube search URL
    search_url = f'https://www.youtube.com/results?search_query={query}'

    # Send a GET request to the search URL
    response = requests.get(search_url)

    if response.status_code == 200:
        # Parse HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the elements containing video views
        view_elements = soup.select('.yt-simple-endpoint.style-scope.ytd-video-meta-block span')

        # Extract video views from the elements
        video_views = [element.get_text() for element in view_elements[:num_views]]

        return video_views

    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

# Example usage: Extract video views for the query 'python tutorial'
video_views = extract_video_views('python tutorial', num_views=5)

if video_views:
    print("First five video views:")
    for idx, views in enumerate(video_views, start=1):
        print(f"{idx}. {views}")
````
